```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow_probability as tfp
tfd = tfp.distributions
tfb = tfp.bijectors
```

```python
dist = tfd.Binomial(total_count=10., probs=1/6)
n_dollar = np.linspace(0, 10, 11, dtype=np.uint8)
plt.stem(n_dollar, dist.prob(n_dollar))
plt.xlabel('Number of $ sign rolls observed')
plt.ylabel('Probability')
_ = plt.show()
```

```python
n_dollar_faces = [0,1,2,3,4,5,6]
dist = tfd.Binomial(total_count=10, probs=np.asarray(n_dollar_faces)/6)
plt.stem(n_dollar_faces, dist.prob(2)/np.sum(dist.prob(2)))  # NB: normalize likelihood 
plt.xlabel('Number of faces with $ signs')
plt.ylabel('Probability of observing 2 $ rolls')
_ = plt.show()
```
