```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data_dir = os.path.expanduser('/home/mattmcd/Work/Data/doing_data_science/dds_datasets/')
```

```python
df = pd.read_csv(
    os.path.join(data_dir, 'nyt1.csv')
)
df.rename(columns={c: c.lower() for c in df.columns}, inplace=True)
df['ctr'] = df['clicks']/df['impressions']
df.head()
```

```python
df['age_group'] = pd.cut(
    df['age'], 
    bins=[-np.inf, 18, 24, 34, 44, 54, 65, np.inf], 
    labels=['<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
)
df.head()
```

```python
df.groupby('age_group')[['impressions', 'clicks', 'ctr']].describe()
```

Lots of users who don't click - add an engagement column for no_imp, imp, clicks

```python
df['engagement'] = pd.Categorical(
    np.where(df['impressions'] == 0, 'NoImp', np.where(df['clicks'] == 0, 'Imp', 'Clicks')),
    ordered=True, categories=['NoImp', 'Imp', 'Clicks']
)
```

```python
df.head()
```

```python
df.loc[df.engagement == 'Clicks', :].groupby('age_group')[['clicks', 'ctr']].describe()
```
