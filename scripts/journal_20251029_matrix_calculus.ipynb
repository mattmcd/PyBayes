```python
import sympy as sp
```

```python
sp.init_printing(use_latex=True)
```

```python
m, n = sp.symbols('m n', integer=True)
y = sp.MatrixSymbol('y', m, 1)
X = sp.MatrixSymbol('X', m, n)
w = sp.MatrixSymbol('w', n, 1)
l = sp.symbols('lambda', )
```

```python
J = (y - X @ w).T @ (y - X @ w) + l*w.T @ w
```

```python
J
```

```python
J.diff(w)
```

```python
J.diff(w).expand()
```

```python
(X.inv() @ (X.T.inv() @ (J.diff(w)/2))).expand()
```
